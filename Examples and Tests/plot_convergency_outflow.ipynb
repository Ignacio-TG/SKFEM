```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('convergency_outflow_conditions.csv')
```

```python
plt.plot(df['N_elem'], df['Error_traccion_normal'], marker='o')
plt.xscale('log', base=2)
plt.yscale('log', base=10)
plt.xlabel('Number of Elements (log scale)')
plt.ylabel('Error in boundary condition (log scale)')
plt.grid(True, which="both", ls="--")
plt.show()
```

```python
# Calculate the slope in log-log scale
log_n_elem = np.log(df['N_elem'])
log_error  = np.log(df['Error_traccion_normal'])

slope = np.polyfit(log_n_elem, log_error, 1)[0]
order_h = -slope * 2  # Convertir a orden en h

print(f'Ordend de convergencia en N: {-slope:.4f}')
print(f'Orden de convergencia en h: {order_h:.4f}')
```

```python
# Calcular pendiente respecto a N
log_n_elem = np.log(df['N_elem'])
log_error  = np.log(df['Error_traccion_normal'])

slope_N = np.polyfit(log_n_elem, log_error, 1)[0]
order_h = -slope_N * 2  # Convertir a orden en h

print(f'Pendiente respecto a N: {slope_N:.4f}')
print(f'Orden de convergencia en h: {order_h:.4f}')
```
