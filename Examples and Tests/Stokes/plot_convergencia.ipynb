```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_kovasznay = pd.read_csv('data/convergencia_stokes_sincos.csv')

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].plot(data_kovasznay['N'], data_kovasznay['err_u_L2'], marker='o')
axs[0, 0].set_ylabel('Error U (L2)')
axs[0, 0].set_yscale('log')
axs[0, 0].set_xscale('log', base=2)

axs[0, 1].plot(data_kovasznay['N'], data_kovasznay['err_u_H1_semi'], marker='o')
axs[0, 1].set_ylabel('Error U (H1 semi)')
axs[0, 1].set_yscale('log')
axs[0, 1].set_xscale('log', base=2)

axs[1, 0].plot(data_kovasznay['N'], data_kovasznay['err_p_L2'], marker='o')
axs[1, 0].set_xlabel('N elements')
axs[1, 0].set_ylabel('Error P (L2)')
axs[1, 0].set_yscale('log')
axs[1, 0].set_xscale('log', base=2)

axs[1, 1].plot(data_kovasznay['N'], data_kovasznay['err_p_H1_semi'], marker='o')
axs[1, 1].set_xlabel('N elements')
axs[1, 1].set_ylabel('Error P (H1 semi)')
axs[1, 1].set_yscale('log')
axs[1, 1].set_xscale('log', base=2)

plt.suptitle('Test de convergencia - SinCos')
plt.tight_layout()
plt.show()
```
