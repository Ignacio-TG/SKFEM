```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Test 1

```python
problem1   = pd.read_csv('data/eigenvalues_problem1.csv')
vals_test1 = [13.086, 23.031, 23.031, 32.053, 38.532, 41.759]
vals_extr1 = [13.0860, 23.0308,23.0308,  32.0443, 38.5353, 41.7588]
```

```python
error= []
for i in range(5):
    error.append(np.abs(np.array(problem1[f'lambda_{i+1}']) - vals_test1[i]))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i in range(5):
    axes[i].plot(problem1['N'], error[i], 'o-', label='SKFEM')
    axes[i].set_xscale('log')
    # axes[i].hlines(vals_test1[i], xmin=4, xmax=128, color='r', linestyle='--', label='Test Value')
    # axes[i].set_xlim(3.8, 140)
    axes[i].set_yscale('log')
    axes[i].legend()
    axes[i].set_title(rf'$\lambda_{{{i+1}}}$ Error')
    axes[i].grid(True, alpha=0.3)

# Ocultar el ltimo subplot (vaco)
axes[5].axis('off')

plt.tight_layout()
plt.show()
```

## Test 2

```python
problem1   = pd.read_csv('data/eigenvalues_problem2.csv')
vals_test1 = [52.3450, 92.1257, 92.1257, 128.2132, 154.1232, 167.0241]
vals_extr1 = [52.3444, 92.1249, 92.1249, 128.2070, 154.1236, 167.0293]
```

```python
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i in range(5):
    axes[i].plot(problem1['N'], problem1[f'lambda_{i+1}'], 'o-', label='SKFEM')
    axes[i].set_xscale('log', base=2)
    axes[i].hlines(vals_test1[i], xmin=4, xmax=128, color='r', linestyle='--', label='Test Value')
    axes[i].set_xlim(3.8, 140)
    axes[i].legend()
    axes[i].set_title(rf'$\lambda_{{{i+1}}}$')
    axes[i].grid(True, alpha=0.3)

# Ocultar el ltimo subplot (vaco)
axes[5].axis('off')

plt.tight_layout()
plt.show()
```
